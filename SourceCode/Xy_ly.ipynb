```python
df = df.dropna(subset=['Mô tả ngắn'])
```

```python
df=df.drop(columns=['Chọn đơn vị tính'])
```

```python
#xóa bỏ kí tự trong ()
df['Tên thuốc'] = df['Tên thuốc'].str.replace(r'\(.*?\)', '', regex=True)
```

```python
df['First_Word'] = df['Tên thuốc'].str.split().str[0]

# Đếm tần suất xuất hiện của từng từ đầu tiên
word_count = df['First_Word'].value_counts()
```

```python
word_count.head(20)
```

```python
# Danh sách các từ khóa
keywords = ['Thuốc', 'Viên', 'Dung', 'Kem', 'Bột', 'Siro', 'Hỗn', 'Gel', 'Dầu', 'Cốm']

# Tạo biểu thức regex để tìm các từ đầu tiên khớp với từ khóa
pattern = r'^(?:' + '|'.join(keywords) + r')\b'

# Xử lý các giá trị NaN trước khi lọc
df = df.dropna(subset=['Tên thuốc'])  # Loại bỏ các dòng có NaN trong cột 'Tên thuốc'

# Lọc các dòng có từ đầu tiên trong 'Tên thuốc' thuộc danh sách từ khóa
filtered_df = df[df['Tên thuốc'].str.contains(pattern, case=True, regex=True)]

# Hiển thị DataFrame đã lọc
print(filtered_df)
```

```python
filtered_df.to_excel("clean_name.xlsx")  # Lưu vào file CSV

```
